```python
newdata = train.rename(columns = {"Reached.on.Time_Y.N":"Reached_on_Time_Y_N"}) 
```

```python

sns.countplot(x = "Prior_purchases", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('Prior_puchase and Reached on-time')
plt.show()
```

```python

sns.countplot(x = "Customer_rating", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('Customer Rating and Reached on-time')
plt.show()
```

```python

sns.countplot(x = "Product_importance", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('Product importance and Reached on-time')
plt.show()
```

```python

sns.countplot(x = "Gender", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('gender and Reached on-time')
plt.show()
```

```python
sns.countplot(x = "Warehouse_block", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N",order = newdata['Warehouse_block'].value_counts().index);
plt.title('warehouse block and Reached on-time')
plt.show()
```

```python
sns.countplot(x = "Mode_of_Shipment", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('Mode of shipment and Reached on-time')
plt.show()
```

```python
sns.countplot(x = "Customer_care_calls", data = newdata, palette = "plasma",hue="Reached_on_Time_Y_N");
plt.title('Customer care calls and Reached on-time')
plt.show()
```
